```python
def get_number_input(prompt):
    """Gets a valid numeric input from the user."""
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("Erro: Entrada inválida. Por favor, digite um número.")

def get_operator_input(prompt):
    """Gets a valid operator input from the user."""
    valid_operators = ['+', '-', '*', '/']
    while True:
        op = input(prompt)
        if op in valid_operators:
            return op
        else:
            print("Erro: Operação inválida. Use '+', '-', '*' ou '/'.")

def add(num1, num2):
    return num1 + num2

def subtract(num1, num2):
    return num1 - num2

def multiply(num1, num2):
    return num1 * num2

def divide(num1, num2):
    if num2 == 0:
        raise ZeroDivisionError("Não é possível dividir por zero!")
    return num1 / num2

def run_calculator():
    """Runs the simple calculator program."""
    print("\n--- Calculadora Profissional ---")
    print("Operações disponíveis: +  -  *  /")

    while True:
        num1 = get_number_input("Digite o primeiro número: ")
        op = get_operator_input("Digite a operação (+, -, *, /): ")
        num2 = get_number_input("Digite o segundo número: ")

        try:
            result = None
            if op == '+':
                result = add(num1, num2)
            elif op == '-':
                result = subtract(num1, num2)
            elif op == '*':
                result = multiply(num1, num2)
            elif op == '/':
                result = divide(num1, num2)

            if result is not None:
                print(f"Resultado: {num1} {op} {num2} = {result}")
        except ZeroDivisionError as e:
            print(f"Erro: {e}")
        except Exception as e:
            print(f"Ocorreu um erro inesperado: {e}")

        another_calculation = input("Deseja fazer outro cálculo? (s/n): ").lower()
        if another_calculation != 's':
            print("Obrigado por usar a calculadora!")
            break

# Call the main calculator function to start the program
run_calculator()
```
